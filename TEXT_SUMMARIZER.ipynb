```python
from transformers import T5Tokenizer, T5ForConditionalGeneration, T5Config
from textwrap3 import wrap
```

```python
text = """The Alps, spanning eight European countries, are a mesmerizing spectacle of nature's grandeur. This majestic mountain range, stretching over 750 miles from France to Slovenia, boasts Mont Blanc as its highest peak at 4,809 meters. Shaped by ancient tectonic forces, the Alps feature diverse climates, from temperate valleys to alpine tundras, nurturing a rich biodiversity. Iconic flora like edelweiss, and fauna like ibex and golden eagles, thrive in these challenging environments. Historically, the Alps have been pivotal, serving as trade routes and migration paths that shaped European culture and politics. Today, they offer a playground for outdoor enthusiasts, with skiing in winter and hiking in summer. Their picturesque alpine villages and rich cultural heritage add to the allure, making the Alps a cherished treasure that embodies nature's beauty and resilience."""
```

```python
model = T5ForConditionalGeneration.from_pretrained('t5-base')
tokenizer = T5Tokenizer.from_pretrained('t5-base')
```

```python
import nltk
nltk.download('punkt')
nltk.download('brown')
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
from nltk.tokenize import sent_tokenize

def postprocesstext (content):
  final=""
  for sent in sent_tokenize(content):
    sent = sent.capitalize()
    final = final +" "+sent
  return final


def summarizer(text,model,tokenizer):
  text = text.strip().replace("\n"," ")
  text = "summarize: "+text
  # print (text)
  max_len = 512
  encoding = tokenizer.encode_plus(text,max_length=max_len, pad_to_max_length=False,truncation=True, return_tensors="pt")

  input_ids, attention_mask = encoding["input_ids"], encoding["attention_mask"]

  # The attention mask is a binary tensor indicating the position of the padded indices so that the model does not attend to them.
  # For the BertTokenizer, 1 indicates a value that should be attended to, while 0 indicates a padded value.

  outs = model.generate(input_ids=input_ids,
                                  attention_mask=attention_mask,
                                  early_stopping=True,
                                  num_beams=3,
                                  num_return_sequences=1,
                                  no_repeat_ngram_size=2,
                                  min_length = 75,
                                  max_length=300)


  dec = [tokenizer.decode(ids,skip_special_tokens=True) for ids in outs]
  summary = dec[0]
  summary = postprocesstext(summary)
  summary= summary.strip()

  return summary


summarized_text = summarizer(text,model,tokenizer)


print ("\nINPUT TEXT FROM USER →")
for wrp in wrap(text, 150):
  print (wrp)
print ("\n")
print ("THE SUMMARIZED TEXT →")
for wrp in wrap(summarized_text, 150):
  print (wrp)
print ("\n")
```
